```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline      

def f2(x, w):
    return (x-w)*x*(x+w)

x = np.linspace(-3, 3, 100)

plt.plot(x, f2(x, 2), color = "black", label = "$w=2$")
plt.plot(x, f2(x, 1), color = "cornflowerblue", label = "$w=1$")

plt.legend(loc="upper left")                # 범례표시
plt.ylim(-15,15)                         # y축 범위
plt.title('$f_2(x)$')                  # 제목
plt.grid(True)                          # 눈금
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    r = 2 * x0**2 + x1**2
    ans = r * np.exp(-r)
    return ans

xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
print(np.round(y,2))        # round() : 소수점 자리 반올림
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

xn = 50
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)
y = np.zeros((len(x0),len(x1)))

for i0 in range(xn):
    for i1 in range(xn):
        y[i0, i1] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0,x1)
plt.figure(1, figsize= (7, 7))
cont = plt.contour(xx0, xx1, y, 5, colors='black')
cont.clabel(fmt='%3.2f',fontsize=8)
plt.xlabel("$x_0$", fontsize=14)
plt.ylabel("$x_1$", fontsize=14)
plt.show()
```

```python
# 함수를 numpy 배열에 적용하기 위해 np.vectorize()
import numpy as np 
matrix = np.arange(1,10).reshape(3,3)
add_100 = lambda i:i+100
vectorized_add_100 = np.vectorize(add_100)
vectorized_add_100(matrix)

find_odd = lambda i : i if i%2==1 else 0
find_even = lambda i : i if i%2==0 else 0       # 삼항연산자
    
vectorized_find_odd = np.vectorize(find_odd)    
vectorized_find_odd(matrix)

vectorized_find_even = np.vectorize(find_even)
vectorized_find_even(matrix)
```

```python
matrix = np.arange(1,10).reshape(3,3)
print("평균: ", np.mean(matrix))
print("분산: ", np.var(matrix))
print("표준편차: ", np.std(matrix))
print(np.std(matrix, ddof=1))                   # 표준편차 구할땐 -1을 해야됨
print("열의 평균 : ", np.mean(matrix, axis=0))
print()

print(matrix)               # 전치행렬
print(matrix.T)
print(matrix.transpose())
np.array([[1,2,3,4,5,6,7]]).T
```

```python
matrix = np.array([[1,1,1],[1,1,10],[1,1,15]])
print(matrix)
np.linalg.matrix_rank(matrix)           # 배열의 행 중 고유행의 개수, 선형대수학에서 matrix의 랭크
```

```python
# 역행렬
matrix = np.array([[1,4],[2,5]])
print(matrix)
inv_matrix = np.linalg.inv(matrix)  # 정방행렬의 역행렬을 구할때 inv()
print(inv_matrix)
```

```python
matrix = np.arange(1,7).reshape(2,3)
inv_matrix = np.linalg.pinv(matrix) # 정방행렬이 아닐때 pinv()
print(inv_matrix)
print()

print(matrix @ inv_matrix)
print()

print(np.round(matrix @ inv_matrix, 0))
```
