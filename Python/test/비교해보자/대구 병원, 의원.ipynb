```python
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
font_path = 'C:/Windows/Fonts/NGULIM.TTF'
font_family = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = font_family
```

```python
df = pd.read_excel('data/6270000_대구광역시_01_01_01_P_병원.xlsx')
df2 = pd.read_excel('data/6270000_대구광역시_01_01_02_P_의원.xlsx')
```

```python
df4 = df3[['상세영업상태명','소재지전체주소','업태구분명']]
df4.head()
```

```python
df5 = df4[df4['상세영업상태명']=='영업중']
df5 = df5.dropna(axis=0)
df5
```

```python
df_hs = df5[df5['업태구분명'].isin(['병원','의원','종합병원'])]
```

```python
df_hs['소재지전체주소']=df_hs['소재지전체주소'].str.lstrip()
df_hs['소재지전체주소']=df_hs['소재지전체주소'].str.replace('동구','동구 ')
df_hs[['시도','구','나머지']]=df_hs['소재지전체주소'].str.split(' ',2,expand=True)
```

```python
df_hs.head()
```

```python
df_hs['count']=1
```

```python
df_hs2 = df_hs.groupby('구').sum()
df_hs2 = df_hs2.sort_values(by='count',ascending=False)
df_hs2
```

```python
df_de = df5[df5['업태구분명'].isin(['치과병원','치과의원'])]
```

```python
df_de['소재지전체주소']=df_de['소재지전체주소'].str.lstrip()
df_de[['시도','구','나머지']]=df_de['소재지전체주소'].str.split(' ',2,expand=True)
df_de
```

```python
df_de['count']=1
```

```python
df_de2 = df_de.groupby('구').sum()
df_de2 = df_de2.sort_values(by='count',ascending=False)
df_de2
```

```python
pop1906 = pd.read_csv('data/2019.06.30.UTF8.csv',thousands=',')
```

```python
dfx = pop1906.copy()
drop_list = []  
check_list = ['총계', '행정구역', ' 남', ' 여'] 
for chk in check_list:  
    for item in dfx.columns :  
        if  chk in item:  
            drop_list.append(item)  
drop_list = list(set(drop_list))
dfx3 = dfx.drop(drop_list,axis=1)
dfx3['인구'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
dfx3['인구'] = np.where(dfx3['인구'] >= 70, 70, dfx3.인구)
dfx3 = dfx3.groupby('인구')
dfx3 = dfx3.sum()
dfx3.columns = dfx3.columns.str.replace(' 계', '')
dfx3
```

```python
dfx3=dfx3.T
```

```python
dfx3['총합']=dfx3.sum(axis='columns').astype(int)
```

```python
dfx4 = dfx3['총합']
```

```python
dfx4
```

```python
result = df_hs2.copy()
result
```

```python
result=result.rename(columns={'count':'병원수'})
result['치과수']=df_de2
result['인구수']=dfx4
result
```

```python
result['병원비율']=result['병원수']/(result['인구수']//1000)
result['치과비율']=result['치과수']/(result['인구수']//1000)
```

```python
result
```

```python

labels = result.index.tolist()
hs_means = result['병원수'].values.tolist()
de_means = result['치과수'].values.tolist()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, hs_means, width, label='병원')
rects2 = ax.bar(x + width/2, de_means, width, label='치과')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('수')
ax.set_title('병원과 치과 수')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0.5),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python

labels = result.index.tolist()
hs_means = result['병원비율'].values.tolist()
de_means = result['치과비율'].values.tolist()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, hs_means, width, label='병원')
rects2 = ax.bar(x + width/2, de_means, width, label='치과')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('비율')
ax.set_title('병원과 치과 비율')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('%.1f'%(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 0.5),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
