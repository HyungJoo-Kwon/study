```python
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
font_path = 'C:/Windows/Fonts/NGULIM.TTF'
font_family = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = font_family
```

```python
df1 = pd.read_csv("data/아파트(매매)__실거래가_수성구.csv",encoding='euc-kr',skiprows=15,thousands=',')
df1.head()
```

```python
df2 = pd.read_csv("data/아파트(매매)__실거래가_달서구.csv",encoding='euc-kr',skiprows=15,thousands=',')
df2.head()
```

```python
df3 = df1.copy()
```

```python
#2-1
df3['전용면적(㎡)']=df3['전용면적(㎡)'].astype('int')
df3.head()
```

```python
#2-2
df3 = df3.sort_values('전용면적(㎡)',ascending=False)
df4 = df3.groupby('전용면적(㎡)')
df4 = df4.count()
df4 = df4.sort_values('시군구',ascending=False)
#2-2
df4[df4['시군구']==df4['시군구'].max()]
```

```python
#달서구데이터
df32 = df2.copy()
df32['전용면적(㎡)']=df32['전용면적(㎡)'].astype('int')
df32 = df32.sort_values('전용면적(㎡)',ascending=False)
df42 = df32.groupby('전용면적(㎡)')
df42 = df42.count()
df42 = df42.sort_values('시군구',ascending=False)
df42[df42['시군구']==df42['시군구'].max()]
```
