```python
# 2. 5-1)
susung1_top7 = susung1.sort_values(by="거래금액(만원)", ascending=False)[0:7]
susung1_top7_value = susung1_top7["거래금액(만원)"].tolist()
susung1_top7_index = susung1_top7.index.tolist()
```

```python
labels = np.ravel(susung1_top7_value)
x = np.arange(len(labels))  # the label locations
width = 0.6  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 10)) #fig, ax = plt.subplots(TOTAL_X,TOTAL_Y,figsize=(8, 5))

plt.grid(b=True, which='both', axis='both')
rects1 = ax.bar(susung1_top7_index, np.ravel(labels), width, align='center', alpha=0.5, label='평균거래금액')
rects1[susung1_top7_value.index(max(susung1_top7_value))].set_color('r')


ax.set_title("수성구 아파트 단지별 평균 거래금액 top7")
ax.set_xticks(x)
ax.set_xticklabels(susung1_top7_index)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
fig.tight_layout()

plt.show()
```
