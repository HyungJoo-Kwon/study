```python
df = df[df["영업상태명"] != "폐업"]
df
```

```python
df1 = pd.DataFrame(df, columns=["소재지전체주소", "사업장명"])
df1["구"] = df1["소재지전체주소"].str[6:9]
```

```python
df1["chicken"] = "0"

df1.loc[df1["사업장명"].str.contains("치킨"), "chicken"] = 1
df1.loc[df1["사업장명"].str.contains("통닭"), "chicken"] = 1 
df1.loc[df1["사업장명"].str.contains("호프"), "chicken"] = 1 
```

```python
df1
```

```python
df1["chicken"] = df1["chicken"].astype(int)
```

```python
result = df1.groupby('구').chicken.agg(["sum"])
result
```
