### 1. 10살 단위로 인구를 정리한다.

```python
import numpy as np
import pandas as pd
```

```python
resident_01 = pd.read_csv('2018.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv')
resident_01.head()
type(resident_01)
```

```python
df = resident_01.copy()
```

```python
drop_list = []
check_list = ['총계', '구분', ' 남', ' 여'] # 남구 삭제 방지를 위해 '남' 앞에 공백 추가 
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
#drop_list
```

```python
df1 = df.drop(drop_list, axis=1)
df1
```

```python
df1['인구0'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
df1 = df1.groupby('인구0')
df1 = df1.sum()
df1
```

### 2. 70대 이상은 하나로 합친다.

```python
df1['인구'] = df1.index
df1['인구'] = np.where(df1['인구'] >= 70, 70, df1.인구)
df1 = df1.groupby('인구')
df1 = df1.sum()
df1
```

### 4. 다음조건 찾기
- 남성인구 > 여성인구
- 인구수 > 20만

```python
import pandas as pd 
import numpy as np
first = pd.read_csv("data/2018.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv")
second = pd.read_csv("data/2018.12.31.기준 대구광역시 주민등록인구 현황_UTF8.csv")
third = pd.read_csv("data/2019.06.30.기준 대구광역시 주민등록인구 현황_UTF8.csv", thousands=',')    # 컴마로 된 거 인식
```

```python
df = third.copy()
```

```python
#공통제거
drop_list = []
check_list = ['총계', '구분',]
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
```

```python
df = df.drop(drop_list, axis=1)
df
```

```python
df['인구'] = (pd.DataFrame(range(0,101))/10).astype(int)*10
df['인구'] = np.where(df['인구'] >= 70, 70, df.인구)
df = df.groupby('인구')
df = df.sum()
df = df.T
df['총합'] = df.sum(axis='columns').astype(int)
df = df.T
df
```

```python
#남자만 뽑기
drop_list = []
check_list = [' 계', ' 여']
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
#drop_list
```

```python
ex4_df1_M = df.drop(drop_list, axis=1)
ex4_df1_M = ex4_df1_M.T
ex4_df1_M
```

```python
#여자만 뽑기
drop_list = []
check_list = [' 계', ' 남']
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
#drop_list
```

```python
ex4_df1_F = df.drop(drop_list, axis=1)
ex4_df1_F = ex4_df1_F.T
ex4_df1_F 
```

```python
#합계만 뽑기
drop_list = []
check_list = [' 남', ' 여']
for chk in check_list:
    for item in df.columns :
        if  chk in item:
            drop_list.append(item)
drop_list = list(set(drop_list)) #  중복 제거를 위해 set을 사용
#drop_list
```

```python
ex4_df1_T = df.drop(drop_list, axis=1)
ex4_df1_T = ex4_df1_T.T
ex4_df1_T
```

```python
ex4_df1_T['남초여부'] = (ex4_df1_M['총합'].values > ex4_df1_F['총합'].values) 
# result = (ex4_df1_M['총합'].values > ex4_df1_F['총합'].values) > 0
# index = ex4_df1_T.index
# obj = pd.Series(result, index=index)
# ex4_df1_T['남초여부'] = obj
ex4_df1_T
```

```python
case1 = ex4_df1_T['남초여부'] == True 
case2 = ex4_df1_T['총합']> 200000
case = case1 & case2
ex4_df1_T['결과'] = case
ex4_df1_T
```
